```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv('dataset.csv')
df.head(5)
```

```python
# Create scatter plot with mode differentiation
fig = px.scatter(df, x = 'key', y = 'valence', color = 'mode', 
                 title = 'Key vs. Valence (by Mode)',
                 labels={'key': 'Key', 'valence': 'Valence', 'mode': 'Mode'},
                 opacity=0.6)

fig.show()
```
