```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
import altair as alt
```

```python
df = pd.read_csv('dataset.csv')

print('Columns:', df.columns.tolist())
```

```python
top_artists = df['artists'].value_counts().nlargest(20).index
df_top = df[df['artists'].isin(top_artists)]

sns.barplot(x = 'artists', y = 'popularity', hue = 'explicit', data = df_top)
plt.xticks(rotation = 90)
plt.title('Popularity by Artists (Top 20), Split by Explicit Hue')
plt.show()
```
