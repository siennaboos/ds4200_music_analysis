```python
import pandas as pd
import altair as alt
```

```python
df = pd.read_csv('dataset.csv')

# Filter for Chuck Berry and make an explicit copy
chuck_df = df[df['artists'].str.contains('Chuck Berry', case=False, na=False)].copy()

# Add new column safely
chuck_df['duration_min'] = chuck_df['duration_ms'] / 60000

chuck_df.head()
```

```python
# Base scatter plot: compare Energy vs Danceability
scatter = (
    alt.Chart(chuck_df)
    .mark_circle(size=100, opacity=0.7)
    .encode(
        x=alt.X('danceability', title='Danceability'),
        y=alt.Y('energy', title='Energy'),
        color=alt.Color('valence', scale=alt.Scale(scheme='viridis'), title='Valence (Mood)'),
        size=alt.Size('popularity', title='Popularity'),
        tooltip=['track_name', 'album_name', 'popularity', 'duration_min', 'valence', 'tempo']
    )
    .properties(
        title='Chuck Berry Songs: Energy vs Danceability (Colored by Valence)',
        width=600,
        height=400
    )
)
scatter
```
